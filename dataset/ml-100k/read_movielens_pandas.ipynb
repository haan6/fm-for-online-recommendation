```python
import numpy as np
from scipy.sparse import coo_matrix
import scipy.io as sio
import matplotlib.pyplot as plt
from datetime import timezone, timedelta, datetime
import pandas as pd

import matplotlib
matplotlib.use('Agg') # Must be before importing matplotlib.pyplot or pylab!
import matplotlib.pyplot as plt
```

```python
#file_path_name = './ub.base'
file_path_name = './ua.base'
# r_cols
r_cols = ['user_id', 'movie_id', 'rating', 'unix_timestamp']
#userId,movieId,rating,timestamp
users = pd.read_csv(file_path_name, sep='\t', names=r_cols,encoding='latin-1')
```

```python
#sorted_data
users2 = users.sort_values(["unix_timestamp"])
```

```python
unsorted_time = np.array([ datetime.utcfromtimestamp(current_utc) for current_utc in users['unix_timestamp'].values])
sorted_time = np.array([ datetime.utcfromtimestamp(current_utc) for current_utc in users2['unix_timestamp'].values])
```

```python
tosee = 1000
fig_size = (10,5)
fig = plt.figure(figsize=fig_size)
fig.add_subplot(1,2,1)
plt.plot(unsorted_time[:tosee],users['rating'].values[:tosee],'.')
plt.xticks(rotation = 70)
plt.title('time_unsorted_rating',fontsize = 15)
plt.gcf().autofmt_xdate()

fig.add_subplot(1,2,2)
plt.plot(sorted_time[:tosee],users2['rating'].values[:tosee],'.')
plt.gcf().autofmt_xdate()
plt.xticks(rotation = 70)
plt.title('time_sorted_rating',fontsize = 15)
plt.show()
```

```python
save_path = './Data_Analysis/'
```

```python
fig_size2 = (8,3)
fig2 = plt.figure(figsize = fig_size2)
#fig2.add_subplot(2,1,1)
plt.plot(sorted_time[:tosee],users2['user_id'].values[:tosee],'r.')
plt.gcf().autofmt_xdate
plt.xticks(rotation=70)
plt.title('time_sorted_user_id',fontsize = 15)
plt.ylabel('user_id',fontsize = 15)
plt.savefig(save_path + 'user_id', bbox_inches='tight')

#fig2.add_subplot(2,1,2)
fig2 = plt.figure(figsize = fig_size2)
plt.plot(sorted_time[:tosee],users2['movie_id'].values[:tosee],'r.')
plt.gcf().autofmt_xdate
plt.ylabel('movie_id',fontsize = 15)
plt.title('time_sorted_movie_id',fontsize = 15)
plt.xticks(rotation=70)
plt.savefig(save_path + 'movie_id', bbox_inches='tight')


fig3 = plt.figure(figsize = fig_size2)
plt.plot(sorted_time[:tosee],users2['rating'].values[:tosee],'.')
plt.ylabel('rating',fontsize = 15)
#plt.gcf().autofmt_xdate()
plt.xticks(rotation=70)
plt.ylim([1-0.1,5.1])

plt.title('time_sorted_rate_id',fontsize = 15)
plt.savefig(save_path + 'rate_id', bbox_inches='tight')


plt.show()



#plt.xticks()

```

```python
downsampling_rate = 20


fig_size2 = (8,3)
fig2 = plt.figure(figsize = fig_size2)
#fig2.add_subplot(2,1,1)
plt.plot(sorted_time[::downsampling_rate][:tosee],users2['user_id'].values[::downsampling_rate][:tosee],'r.')
plt.gcf().autofmt_xdate
plt.xticks(rotation=70)
plt.title('downsampling' +str(downsampling_rate) + '_time_sorted_user_id',fontsize = 15)
plt.ylabel('user_id',fontsize = 15)
plt.savefig(save_path + 'downsampling' +str(downsampling_rate) +'user_id', bbox_inches='tight')


#fig2.add_subplot(2,1,2)
fig2 = plt.figure(figsize = fig_size2)
plt.plot(sorted_time[::downsampling_rate][:tosee],users2['movie_id'].values[::downsampling_rate][:tosee],'r.')
plt.gcf().autofmt_xdate
plt.ylabel('movie_id',fontsize = 15)
plt.title('downsampling' +str(downsampling_rate) + '_time_sorted_movie_id',fontsize = 15)
plt.xticks(rotation=70)
plt.savefig(save_path + 'downsampling' +str(downsampling_rate) +'movie_id', bbox_inches='tight')


fig3 = plt.figure(figsize = fig_size2)
plt.plot(sorted_time[::downsampling_rate][:tosee],users2['rating'].values[::downsampling_rate][:tosee],'.')
plt.ylabel('rating',fontsize = 15)
#plt.gcf().autofmt_xdate()
plt.xticks(rotation=70)
plt.ylim([1-0.1,5.1])
plt.title('downsampling' +str(downsampling_rate) + '_time_sorted_rate_id',fontsize = 15)
plt.savefig(save_path + 'downsampling' +str(downsampling_rate) +'rate_id', bbox_inches='tight')


plt.show()



#plt.xticks()

```
